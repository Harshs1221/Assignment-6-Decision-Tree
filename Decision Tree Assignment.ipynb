```python
#Q1. What is a Decision Tree, and how does it work?
"""
Ans1. A decision tree is a powerfull machine learning algorithm used for classification and regression tasks.
      A decision tree is like a flowchart that helps make decision by asking series of simple yes/no question.
      Image you're deciding what to wear:
      1. Is it raining -
              If yes, wear a raincoat
              If NP, move to the next question.
      2. Is it cold?
              If yes, wear a jacket.
              If no, wear a T-shirt.
        Each question leads to a decision, and this is how decision tree works- breaking a big decision imto smaller
        steps
"""
```

```python
#Q2.  What are impurity measures in Decision Trees?
"""
Ans2. Impurity measures in Decision Tree help determine how well a dataset is split at each node.
      They quantify the homogeneity of the data-lower the impurity means better classification.

      Impurity Measures Types:
      1. Gini -
            Measures the probability of incorrect classification. A lower gini value means a purer node.
      2. Entropy:
            Based on information theory, it calculates the randomness in a dataset. Lower entropy means better splits.
      3. Classification Error:
            Simply counts the fraction of misclassified samples, but is less commonly used.
      4. Information Gain:
            The reduction in entropy after a split , helping choose the best feature for division.
"""
```

```python
#Q5.  What is Information Gain, and how is it used in Decision Trees?
"""
Ans5. Information gain is a measure used in decision tree to determine the best feature for  splitting the data
      at each node. it quantifies the reduction in entropy after a split.
      It will be used for splitting the feature in decision tree, which feature has the maximum information gain, that
      will be used.
"""
```

```python
#Q6. What is the difference between Gini Impurity and Entropy?
"""
Ans6. The differences between Gini Impurity and Entropy are:
        1. Mathematical Complexity:
               - Gini is easier to compute.
               - Entropy more computationally expensive.
        2. Raneg of values:
               - Gini between 0 and 0.5
               - Entropy between 0 and 1
        3. Interpretation:
               - Gini Measures missclassification probability.
               - Entropy Measures randomness or disorder.
        4. Preferred Use case:
               - Gini works well for larger datasets
               - Entropy more usefull for imbalanced datasets.
"""
```

```python
#Q7.  What is the mathematical explanation behind Decision Trees?
"""
Ans7. A decision tree is a supervised leanring algorithm uses a tree like structure to make
      decisions.
      It works by recursively splitting data based on certain conditioins, aiming to maximize the 
      information gain or minimize impurity.
        Formula of entropy is:      
                    ùêª(ùëÜ)=   ‚àí‚àëùëñ=1ùëõ-1 ùëùùëñ log2(ùëùùëñ)

        G.I = 
            1-‚àëi=0 to n-1 (pi)**2
              where pi = is the probablitity of a class in the node
              n = is the total number of classes.

        Information Gain = 
                        impurity(pre-split) - impurity(post-split)      

     - Splitting criteria:
             The algorithm selects the feature that maximizes information gain and
             minimizes gini impuroty.
     - recursive Paririoning:
             The tree continues splitting until a stopping condition is met (e.g., pure nodes).
"""
```

```python
#Q8. What is Pre-Pruning in Decision Trees?
"""
Ans8. Pre-Pruning is used to control the growth of the tree and preventing overfitting.
      It works bt applying constraints before the tree fully develops.
      - Hyperparameter tuning is used to select the best parameter.
      Some common pre pruning methods include:
      1. Setting a maximum depth:
                  Limits how deep the tree can grow.
      2. Minimums samples per split:
                  Ensures each splits has enough data points.
      3. Minimum leaf size:
                  Prevents tiny branches with very few samples
        
"""
```

```python
#Q9. What is Post-Pruning in Decision Trees?
"""
Ans9. Post pruning in decision tree is a technique is used to simplify the tree after it has been
      fully grown.
      Post pruning allows the tree from growing too deep and then removes unnecessary branches to
      improve genralization.
"""
```

```python
#Q10.  What is the difference between Pre-Pruning and Post-Pruning?
"""
Ans10. Difference between Pre-Pruning and Post-Pruning are:

        Pre-Pruning:
            1. Pre-Pruning happens before the tree is fully grown.
               Post-Pruning happens after the tree is fully grown.
            2. Pre-Pruning stops tree growth early using constraints(e.g., max_depth, min sample per split)
               Post-Pruning grows the full tree first, then remove the unnecessary branches.
            3. Prevent excessive complexity from the start.
               Simplify the tree after training.
            4. Risk- Might stop too early and important patterns.
               Post-pruning might remove useful branches if not done carefully. 
            
"""
```

```python
#Q11.  What is a Decision Tree Regressor?
"""
Ans11. Decision Tree Regressor predicts continous numerical values instead of categories.
       It works by splitting the data into smaller regions based on feature values, similar to
       how a decision tree classifier works.
       In decision tree regressor we use variance instead of gini and entropy and variance
       reduction instaed of Imformation Gain.
"""
```

```python
#Q12.  What are the advantages and disadvantages of Decision Trees?
"""
Ans12. Advantages:
        1. Easy to understand & Interpret:
                Decision trees are intutive and can be visualized.
        2. Handles both Numerical & Categorical Data:
                Works well with different types of data.
        3. No Need of Feature Scaling:
                Unlike SVM aor logistic regression, decision tree don't require normalization.
        4. Can Handle Missing Values:
                Some decision tree algorithms can work if the data is incomplete.
        5. Non-parametric:
                No assumption about data distribution are needed.
        6. Captures Non-Linear Relationships:
                Can model complex patterns in data.

                
        Disadvantages:
        1. Prone to Overfitting:
                If the tree grows too deep, it may memorize the training data instead of generalizing.
        2. Sensitive to Noisy Data:
                Small changes in data can lead to completely different trees.
        3. Greedy Algorithm:
                Decision tree use a greedy approach, which may not always find the best spkit.
        4. Computationally expensive for larger datasets.
        5. Less accurate compared to Ensembles methods.

"""
```

```python
#Q13.  How does a Decision Tree handle missing values?
"""
Ans13. Decision tree handle missing value in several ways:
        1. Ignoring missing values during spltting:
                Some algorithms, like CART, can ignore missing values when chossing the
                best split.
        2. Assigning the probabilities to Missing Values
                Instead of removing missing values, decison tree can assign probabilities 
                based on existing data patterns.
        3. Using Surrogate Splits:
                If a primary feature has missing values, the tree finds and alternative
                (surrogate) feature that closely mimics the original split.
        4. Replacing Missing value with Mean/Median/Mode
                Some implementations replace missing values with the most common or
                average value in the dataset.
"""
```

```python
#Q14.  How does a Decision Tree handle categorical features?
"""
Ans14. Decision tree can handle categorical features in different ways:
        1. Label encoding
            - Convert the categorical into numerical labels.(red- 1, blue -2)
            - Works well for ordinal categories.
        2. One Hot Encoding
            - Coverts categories into binary columns (e.g, "Red", 'Blue', 'Green' ->
                [1,0,0], [0,1,0],[0,0,1])
            - Prevent false ranking issues
        3. Decision Tree can handle categorical data directly like CART can split categorical
           features without encoding.
           For example: feature has Sunny, Rainy, Cloudy it splits directly (Sunny, Not sunny)
"""
```

```python
clf = DecisionTreeClassifier(random_state=1)
```

```python
clf.fit(X_train, y_train)
```

```python
#find optimal value for CCP by checking cost complexity pruning
path = clf.cost_complexity_pruning_path(X_train, y_train)
```

```python
#exclude the last ccp as it prune everything
ccp_alpha = path.ccp_alphas[:-1]
```

```python
train_acc = []
test_Acc = []
```

```python
for alpha in ccp_alpha:
    clf_pruned = DecisionTreeClassifier(random_state=1 , ccp_alpha = alpha)
    clf_pruned.fit(X_test,y_test)
```

```python
#calculate the accuracy in training and testing sets
train_acc.append(accuracy_score(y_train, clf_pruned.predict(X_train)))
test_Acc.append(accuracy_score(y_test, clf_pruned.predict(X_test)))
```

```python
plt.plot(ccp_alpha[:len(train_acc)], train_acc, marker="o")
plt.plot(ccp_alpha[:len(test_Acc)], test_Acc, marker="o")
plt.show()
```
